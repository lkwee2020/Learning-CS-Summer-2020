Now that we have tested data with Hawaii data, we can do it with the full data set.
So I will start by reducing the covid data to only the latest data.

```python
covid_latest = covid_data[covid_data.date == '2020-07-12']
```

```python
covid_latest
```

```python
county_data = county_data[county_data.Rank.isna() == False]
```

```python
combined_data = pd.merge(county_data, covid_latest, left_on = 'County_or_county-equivalent', right_on = 'county', how = 'right')
```

```python
combined_data
```

```python
combined_data = combined_data[combined_data.Rank.isna() == False]
```

```python
combined_data = combined_data[combined_data.duplicated(subset = 'County_or_county-equivalent') == False]
```

```python
combined_data
```

Now that the data has been combined into one csv file, we can graph it.

```python
#plt.plot(combined_data.Per_capitaincome, combined_data.cases)
```

Running this cell takes long time because there is so much data. Let's only graph 1,000 rows.

```python
less_data = combined_data.iloc[0:1000]
```

```python
less_data
```

```python
plt.plot(less_data.Per_capitaincome, less_data.cases, 'o')
```

```python
show = less_data[less_data.cases > 120000]
```

```python
show
```

It is interesting that the number of cases is not correlated with the per capita income for each county. Possible reasons why: 1) different counties are experiencing different stages of the pandemic. 2) counties have varying methods of testing population for coronavirus and some might have more access to testing kits etc than others, making the data less certain. 3) counties have varying policies on masks and social distancing requirements. 

Now I will graph number of deaths versus per capita income.

```python
plt.plot(less_data.Per_capitaincome, less_data.deaths, 'o')
```

Again, the data suggest that percapita inocme is not correlated with the number of deaths due to covid19 in counties accross the US.
